```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Wine_Quality_Data.csv")
df
```

**Hypothesis test**

```python
sample=df.chlorides.sample(frac=0.05)
sample
```

```python
for i in range(10):
  sample=df.chlorides.sample(frac=0.05)
  t_value,p_value=stats.ttest_1samp(sample,df.chlorides.mean())
  print(t_value,p_value)
  if (p_value<0.05):
    print("H0 is rejected,Ha is accepted,independent, data is distributed properly")
  else:
    print("H0 is accepted,Ha is rejected, dependent, data is distributed properly")
```

**One sample t test**

```python
df
```

```python
sample1=df.sulphates.sample(frac=0.05)
sample2=df.alcohol.sample(frac=0.05)
```

```python
t_test,p_value=stats.ttest_ind(sample1,sample2)
t_test,p_value
```

```python
for i in range(10):
  sample1=df.sulphates.sample(frac=0.05)
  sample2=df.alcohol.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)
  print(t_test,p_value)
  if (p_value<0.05):
    print("H0 is rejected,Ha is accepted,independent, data is distributed properly")
  else:
    print("H0 is accepted,Ha is rejected, dependent, data is distributed properly")
```
